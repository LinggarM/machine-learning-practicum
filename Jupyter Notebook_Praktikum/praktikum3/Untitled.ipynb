```python
%matplotlib inline
from copy import deepcopy
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
# Import dataset
# Sesuaikan dengan lokasi file xclara.csv di local komputer
data = pd.read_csv('xclara.csv')
print(data.shape)
(3000, 2)
print(data.head(20))
```

```python
# Plot dataset
f1 = data['V1'].values
f2 = data['V2'].values
X = np.array(list(zip(f1, f2)))
plt.scatter(f1, f2, c='black', s=7)
```

```python
from sklearn.cluster import KMeans
# Menentukan jumlah cluster
kmeans = KMeans(n_clusters=3)
# Fitting input data
kmeans_hasil = kmeans.fit(X)
# Mendapatkan cluster labels
labels = kmeans_hasil.predict(X)
# Mendapatkan nilai centroid
C = kmeans_hasil.cluster_centers_
# Mencetak nilai centroid
print(C)
```

```python
plt.scatter(X[:, 0], X[:, 1], s=7, c=labels)
plt.scatter(C[:, 0], C[:, 1], marker='*', s=200, c='#050505')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
plt.rcParams['figure.figsize'] = (16, 9)
# Men-generate dataset yang terkelompok dalam 4 cluster
X, y = make_blobs(n_samples=800, n_features=3, centers=4)
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2])
```

```python
# Initializing KMeans
kmeans = KMeans(n_clusters=4)
# Fitting with inputs
kmeans = kmeans.fit(X)
# Predicting the clusters
labels = kmeans.predict(X)
# Getting the cluster centers
C = kmeans.cluster_centers_
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y)
ax.scatter(C[:, 0], C[:, 1], C[:, 2], marker='*', c='#050505', s=1000)
```

```python
for k in range (1, 10):
    # Menentukan jumlah cluster
    kmeans = KMeans(n_clusters=k, random_state=1)
    # Fitting input data
    kmeans_hasil = kmeans.fit(X)
    # Mendapatkan cluster labels
    labels = kmeans_hasil.predict(X)
    # Menghitung jumlahan jarak antara setiap sampel dengan cluster centroid-nya (SSE)
    interia = kmeans.inertia_
    print ("k:",k, " cost:", interia)
print ("")
```

```python
from sklearn.metrics.cluster import silhouette_score
silhouette_score(X, labels)
```
