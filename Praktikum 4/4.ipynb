```python
#Menggunakan Elbow Method : agar hasil clustering nya optimal

%matplotlib inline
#%matplotlib inline sets the beckend of matplotlib to the 'inline' beckend
import matplotlib.pyplot as plt
#Matplotlib is a ploting library for the Python programming language and its numerical mathematics extension NumPy
plt.style.use('seaborn-whitegrid')
#Seaborn style on matplotlib plot, menentukan jenis graph. untuk jenis graph lain bisa dilihat di https://python-graph-gallery.com/199-matplotlib-style-sheets/
import numpy as np
from sklearn.cluster import KMeans
```

```python
# membuat data set
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T
#mengambil data dari fitur data iris
plt.plot()
plt.scatter(features[0], features[1]) #2 fitur yang akan dipakai
plt.show()
#catatan: jika ingin mencetak/mengetahui data iris, di baris paling bawah bisa diketik iris. (atau iris.data, atau iris.target_names)
iris.target_names
```

```python
X = np.array(list(zip(features[0], features[1]))).reshape(len(features[1]), 2)
SSE = []
K = range(1,10)
for k in K:
  kmeanModel = KMeans(n_clusters=k).fit(X)
  kmeanModel.fit(X)
  SSE.append(kmeanModel.inertia_)
```

```python
plt.plot(K, SSE, 'bx-')
plt.xlabel('k')
plt.ylabel('SSE')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
y_pred = KMeans(n_clusters=3).fit_predict(X)
plt.plot
LABEL_COLOR_MAP = {0 : 'r',
                   1 : 'g',
                   2 : 'b'
                   }
label_color = [LABEL_COLOR_MAP[l] for l in y_pred]
plt.scatter(features[0], features[1], c = label_color)
plt.title("K=3")
plt.show()
```

```python
figure, ax = plt.subplots(2,3)
K=range(1,7)
for k in K :
  if (k<4) :
    row = 0
    column = k-1
  else :
    row = 1
    column = k-4
  kmeanModel = KMeans(n_clusters=k).fit(X)
  y_pred = kmeanModel.fit_predict(X)
  ax[row][column].scatter(features[0], features[1], c=y_pred, cmap='winter_r')
plt.show()
```
